```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the cleaned data
file_path = "/content/data/MachineLearningRating_v3.txt"
df = pd.read_csv(file_path, sep="|")

# Quick check
df.info()
df.describe()
df.head()
```

```python
df["LossRatio"] = df["TotalClaims"] / df["TotalPremium"]

# Inspect
df[["TotalPremium", "TotalClaims", "LossRatio"]].head()
```

```python
# Check for missing values
df.isnull().sum()

# Check duplicates
df.duplicated().sum()
```

```python
# TotalPremium
plt.figure(figsize=(8,5))
sns.histplot(df["TotalPremium"], bins=50, kde=True)
plt.title("Distribution of Total Premium")
plt.show()

# TotalClaims
plt.figure(figsize=(8,5))
sns.histplot(df["TotalClaims"], bins=50, kde=True)
plt.title("Distribution of Total Claims")
plt.show()

# Loss Ratio
plt.figure(figsize=(8,5))
sns.histplot(df["LossRatio"], bins=50, kde=True)
plt.title("Distribution of Loss Ratio")
plt.show()
```

```python
# Province
plt.figure(figsize=(10,5))
sns.boxplot(x="Province", y="LossRatio", data=df)
plt.xticks(rotation=45)
plt.title("Loss Ratio by Province")
plt.show()

# Gender
plt.figure(figsize=(6,4))
sns.boxplot(x="Gender", y="LossRatio", data=df)
plt.title("Loss Ratio by Gender")
plt.show()

# Vehicle Type
plt.figure(figsize=(10,5))
sns.boxplot(x="VehicleType", y="LossRatio", data=df)
plt.xticks(rotation=45)
plt.title("Loss Ratio by Vehicle Type")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df["TotalPremium"], bins=50, kde=True, color="skyblue")
plt.title("Distribution of Total Premium")
plt.xlabel("Total Premium")
plt.ylabel("Count")
plt.show()
```

```python

province_claims = df.groupby("Province")["TotalClaims"].sum().sort_values(ascending=False)

plt.figure(figsize=(10,5))
sns.barplot(x=province_claims.index, y=province_claims.values, palette="viridis")
plt.title("Total Claims by Province")
plt.xlabel("Province")
plt.ylabel("Total Claims")
plt.xticks(rotation=45)
plt.show()
```

```python
df["TransactionMonth"] = pd.to_datetime(df["TransactionMonth"])

monthly_claims = df.groupby(df["TransactionMonth"].dt.to_period("M"))["TotalClaims"].sum()

plt.figure(figsize=(10,5))
monthly_claims.plot(kind="line", marker="o", color="red")
plt.title("Monthly Total Claims")
plt.xlabel("Month")
plt.ylabel("Total Claims")
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python

for col in df.columns:
    if df[col].dtype == 'object':
        df[col] = df[col].astype(str).str.strip()
        df[col] = pd.to_numeric(df[col], errors='ignore')
```

```python
numeric_df = df.select_dtypes(include=['number'])
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(numeric_df.corr(), annot=False, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python
# Correlation with TotalClaims
claims_corr = numeric_df.corr()['TotalClaims'].sort_values(ascending=False)
print("Correlation with TotalClaims:")
print(claims_corr.head(15))
print("\nNegative correlations:")
print(claims_corr.tail(15))
```

```python
top_pos = claims_corr.drop('TotalClaims').head(10)
top_neg = claims_corr.drop('TotalClaims').tail(10)

plt.figure(figsize=(12,6))
top_pos.plot(kind='bar', title='Top Positive Correlations with TotalClaims')
plt.show()

plt.figure(figsize=(12,6))
top_neg.plot(kind='bar', title='Top Negative Correlations with TotalClaims')
plt.show()
```

```python
# Example: Claim Frequency & Claim Severity calculation
df['ClaimOccurred'] = df['TotalClaims'] > 0
claim_freq = df.groupby('Province')['ClaimOccurred'].mean()  # Frequency
claim_sev = df[df['ClaimOccurred']]['TotalClaims'].groupby(df['Province']).mean()  # Severity

print("Claim Frequency by Province:")
print(claim_freq)
print("Claim Severity by Province:")
print(claim_sev)

# You can then run chi-squared or t-tests for A/B testing:
from scipy.stats import ttest_ind, chi2_contingency

# Example: t-test for Claim Severity between provinces
province_1 = df[df['Province'] == 'Gauteng']['TotalClaims'].dropna()
province_2 = df[df['Province'] == 'Western Cape']['TotalClaims'].dropna()
t_stat, p_val = ttest_ind(province_1, province_2)
print(f"T-test between Gauteng & Western Cape: t={t_stat:.3f}, p={p_val:.3f}")
```
